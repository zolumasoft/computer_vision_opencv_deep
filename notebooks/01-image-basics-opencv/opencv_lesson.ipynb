```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
# cv2.imread always reads even if path to image is wrong so check 
# img type (use type(imge)) before continuing
img = cv2.imread('../images/00-puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)

# Reason for blue color: Matplotlib expects RGB -> Red, Green and Blue
# However: OpenCV reads as BGR -> Blue, Green and Red
```

```python
# Convert BGR to RGC with OpenCV cvtColor
fixed_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fixed_img)
```

```python
# Read image as gray scale
img_gray = cv2.imread('../images/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
# Need to add cmap to show gray
# cmap -> color map
plt.imshow(img_gray, cmap='gray')
```

```python
# Resizing image to the correct size
# Check image size: image.shape -> example (1300, 1950)
new_img = cv2.resize(fixed_img, (1000, 400))
```

```python
plt.imshow(new_img)
```

```python
# Reduce image by 50%
w_ratio = 0.5
h_ratio = 0.2 # 20% of original height
```

```python
new_resized_img = cv2.resize(fixed_img, (0, 0), fixed_img, w_ratio, h_ratio)
```

```python
plt.imshow(new_resized_img)
```

```python
# Flipping an image
flipped_img = cv2.flip(fixed_img, 1)
plt.imshow(flipped_img)
```

```python
cv2.imwrite('totally_new.jpg', fixed_img)
```
